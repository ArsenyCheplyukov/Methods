```python
import pandas as pd
import numpy as np
```

```python
with open("field.txt", 'r') as f:
    temperature = [[float(n) for n in line.strip().split(' ')] for line in f]
with open("dT.txt", 'r') as f:
    y_input = np.array([float(x) for x in f.readline().split()])
    x_input = np.linspace(0, len(y_input), len(y_input))
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(
    rows=2, cols=1,
    specs=[[{'type': 'surface'}], [{'type': 'xy'}]]
)

# Generate data
y = np.linspace(1, len(temperature) + 1, len(temperature))
x = np.linspace(1, len(temperature[0]) + 1, len(temperature[0]))
#xGrid, yGrid = np.meshgrid(y, x)

print(len(temperature), len(temperature[0])) 

# adding surfaces to subplots.
fig.add_trace(
    go.Surface(x=x, y=y, z=temperature, colorscale='Viridis', showscale=False),
    row=1, col=1)

fig.add_trace(go.Scatter(x=x_input, y=y_input,
                    mode='lines+markers',
                    name='lines+markers'))

fig.update_layout(
    title_text='3D subplots with different colorscales',
    height=800,
    width=800
)

fig.show()
```
