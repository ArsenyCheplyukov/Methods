```python
import pandas as pd
import numpy as np
from os import listdir
from os.path import isfile, join
import time
```

```python
%matplotlib inline  
from plotly import __version__ 
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objects as go
import cufflinks as cf
init_notebook_mode(connected=True)
cf.go_offline()

folder_name = "./graphs/"

# Generate data
fileslen = len([f for f in listdir(folder_name) if isfile(join(folder_name, f))])

# adding surfaces to subplots.
trace = []
for i in range(fileslen):
    with open(f"{folder_name}T{i}.txt", 'r') as f:
        temperature = ([[float(n) for n in line.strip().split(' ')] for line in f])
    trace.append(go.Surface(z=temperature, colorscale='Viridis', showscale=False))
iplot([trace[0]])
iplot([trace[-1]])
```
